```julia
using PyPlot
using StatsBase
using Distributions
using StatsFuns

```

```julia
N = 100000
z = rand(MvNormal([1.0, 0.0], [1.0 0.5; 0.5 0.5]), N);
x = logistic.(z[1, :]);
y = logistic.(z[2, :]);
h = fit(Histogram, (x, y), closed = :left, nbins = (20, 20));

n_gaps = 5
gap = trunc(Int, maximum(h.weights) / n_gaps)
levels = collect(0:gap:((n_gaps+1)*gap))
level_labels = map(x -> "$(x*10)%", 0:(n_gaps+1))


fig, ax= plt.subplots(1,1, figsize=(7,5), )

# PyPlot.svg(true)
cs = ax.contour(midpoints(h.edges[1]),
             midpoints(h.edges[2]),
             h.weights';
             colors = "black",
             levels = levels)
ax.clabel(cs, fmt = Dict(zip(levels, level_labels)), inline = 1);
```

```julia
N = 100000
z = rand(MvNormal([1.0, 0.0], [1.0 0.5; 0.5 0.5]), N);
x = logistic.(z[1, :]);
y = logistic.(z[2, :]);
h = fit(Histogram, (x, y), closed = :left, nbins = (20, 20));

n_gaps = 5
gap = trunc(Int, maximum(h.weights) / n_gaps)
levels = collect(0:gap:((n_gaps+1)*gap))
level_labels = map(x -> "$(x*10)%", 0:(n_gaps+1))

colors = [plt.cm.jet(i) for i in range(0, stop=0.5, length=n_gaps+1)]

fig, ax= plt.subplots(1,1, figsize=(7,5), )

# PyPlot.svg(true)
cs = ax.contourf(midpoints(h.edges[1]),
             midpoints(h.edges[2]),
             h.weights';
             colors=colors,
             levels = levels)
```
