```julia
using JLD2
using FileIO
using PyPlot
using Statistics
using StatsBase 
using Distributions
using LinearAlgebra
using HDF5
using IntervalSets
using Random, ArraysOfArrays
using ValueShapes
using Measurements
using BAT
using DSP
```

```julia
@load "../../data/sampling_results/Benchmark-10/mcmc-summary-data.jld" data_save

data1 = data_save[1]; # 1 Gaussian model
charges_1 = data1.ev_charge
charges_1n = charges_1 ./ maximum(charges_1)
charges_1n[sortperm(charges_1n)] .= 1:length(charges_1n)
charges_1n = charges_1n ./ maximum(charges_1n)
runind_n1 = data1.ev_index ./ maximum(data1.ev_index)
rind_1 = 1:length(data1.ev_charge)
```

```julia
@load "../../data/sampling_results/Benchmark-5/mcmc-summary-data.jld" data_save

data2 = data_save[1]; # 1 Gaussian model
charges_2 = data2.ev_charge
charges_2n = charges_2 ./ maximum(charges_2)
charges_2n[sortperm(charges_2n)] .= 1:length(charges_2n)
charges_2n = charges_2n ./ maximum(charges_2n)
runind_n2 = data2.ev_index ./ maximum(data2.ev_index);
rind_2 = 1:length(data2.ev_charge)
```

## Transversal Size: 

```julia
σ_x = data1.μx_trsz
σ_y = data1.μy_trsz

σ_x_core = data2.μx_trsz_core
σ_y_core = data2.μy_trsz_core

σ_x_halo = data2.μx_trsz_halo
σ_y_halo = data2.μy_trsz_halo;
```

```julia
xmin, xmax = 0.02, 0.24
xrange=range(xmin, stop = xmax,length=100)
yrange=range(xmin, stop = xmax,length=100)
xvals = [pdf(truncated(Normal(0.2, 0.04), 0.02, 0.25), x) for x in xrange]
yvals = [pdf(truncated(Normal(0.2, 0.04), 0.02, 0.25), y)  for y in yrange]
z = xvals*yvals';
```

```julia
fig, ax= plt.subplots(1,1, figsize=(6,6), )

ax.scatter(σ_x, σ_y, s=10, c=plt.cm.GnBu.(charges_1n), label="Model 1")
ax.scatter(σ_x_core, σ_y_core, s=10, c=plt.cm.YlOrRd.(charges_2n), alpha=0.5, label="Model 2 (Core)")
ax.scatter(σ_x_halo, σ_y_halo,  s=10, c=plt.cm.YlOrRd.(charges_2n), alpha=0.5, label="Model 2 (Halo)")

ax.set_xlabel(L"\sigma_x \; (\mathrm{mm})")
ax.set_ylabel(L"\sigma_y \; (\mathrm{mm})")

ax.legend(loc="upper left")
```

## Angular Spread: 

```julia
σ_x = data1.μx_angspr
σ_y = data1.μy_angspr

σ_x_core = data2.μx_angspr_core
σ_y_core = data2.μy_angspr_core

σ_x_halo = data2.μx_angspr_halo
σ_y_halo = data2.μy_angspr_halo;
```

```julia
xmin, xmax = 0, 10
xrange=range(xmin, stop = xmax,length=300)
yrange=range(xmin, stop = xmax,length=300)
xvals = [pdf(Normal(4, 2), x) for x in xrange]
yvals = [pdf(Normal(4, 2), y)  for y in yrange]
z = xvals*yvals';
```

```julia
fig, ax= plt.subplots(1,1, figsize=(6,6), )

ax.scatter(σ_x, σ_y, s=10, c=plt.cm.GnBu.(charges_1n), label="Model 1")
ax.scatter(σ_x_core, σ_y_core, s=10, c=plt.cm.YlOrRd.(charges_2n), alpha=0.5, label="Model 2 (Core)")
ax.scatter(σ_x_halo, σ_y_halo,  s=10, c=plt.cm.YlOrRd.(charges_2n), alpha=0.5, label="Model 2 (Halo)")

ax.set_xlabel(L"\sigma'_x \; (\mathrm{10^{-5} \cdot rad})")
ax.set_ylabel(L"\sigma'_y \; (\mathrm{10^{-5} \cdot rad})")

ax.legend(loc="upper left")
```

## Emittance: 

```julia
σ_x = data1.μx_emitt
σ_y = data1.μy_emitt

σ_x_core = data2.μx_emitt_core
σ_y_core = data2.μy_emitt_core

σ_x_halo = data2.μx_emitt_halo
σ_y_halo = data2.μy_emitt_halo;
```

```julia
prior_emmitance_x = rand(Normal(0.2, 0.04), 3*10^6) .* rand(Normal(4, 2), 3*10^6) .* 10^-5 .* 1000 .* 426.0
prior_emmitance_y = rand(Normal(0.2, 0.04), 3*10^6) .* rand(Normal(4, 2), 3*10^6) .* 10^-5 .* 1000 .* 426.0

hist_em_tmp = fit(Histogram, (prior_emmitance_x,prior_emmitance_y),nbins=50)
hist_em_tmp = normalize(hist_em_tmp, mode=:pdf);

range_tmp = range(1.2*minimum(hist_em_tmp.weights), stop=maximum(hist_em_tmp.weights), length=10)[2:end]
colors_tmp = [plt.cm.bone_r(i) for i in range(0, stop=0.4, length=length(range_tmp))[2:end]];
```

```julia
fig, ax= plt.subplots(1,1, figsize=(6,6), )

ax.scatter(σ_x, σ_y, s=10, c=plt.cm.GnBu.(charges_1n), label="Model 1")
ax.scatter(σ_x_core, σ_y_core, s=10, c=plt.cm.YlOrRd.(charges_2n), alpha=0.5, label="Model 2 (Core)")
ax.scatter(σ_x_halo, σ_y_halo,  s=10, c=plt.cm.YlOrRd.(charges_2n), alpha=0.5, label="Model 2 (Halo)")

ax.set_xlabel(L"\epsilon_x \; (\mathrm{mm \cdot mrad})")
ax.set_ylabel(L"\epsilon_y \; (\mathrm{mm \cdot mrad})")

ax.legend(loc="upper left")
```
