```julia
function cov_matr(smpl_tmp)
    smp_fl, w = flatview(unshaped.(smpl_tmp.v)), smpl_tmp.weight
    smp_fl = smp_fl ./ std(smp_fl, FrequencyWeights(w), 2)
    smp_fl = smp_fl .- mean(smp_fl, FrequencyWeights(w), dims=2)
    cov_m = cov(smp_fl, FrequencyWeights(w), 2)
    return cov_m
end
```

```julia
function log_lik(; data = summary_data, nevents = n_events)
    
    nevents = n_events
    s_cam = Float64[1.478, 15.026, 23.1150]
    return params -> begin   
        
        ll = 0.0
    
        for i in 1:nevents
            
            x_expected = cos(params.α[i]).*tan(params.θ[i]*10^-5).*s_cam.*10^3
            y_expected = sin(params.α[i]).*tan(params.θ[i]*10^-5).*s_cam.*10^3
            
            x_expected += data.μ_x[i][1] .+ params.x_alignm 
            y_expected += data.μ_y[i][1] .+ params.y_alignm
            
            ll += sum(logpdf.(Normal.(1, params.σ_x), x_expected./data.μ_x[i][2:end] ))
            ll += sum(logpdf.(Normal.(1, params.σ_y), y_expected./data.μ_y[i][2:end] ))
            
#             ll += sum(logpdf.(Normal.(x_expected, params.σ_x), data.μ_x[i][2:end] ))
#             ll += sum(logpdf.(Normal.(y_expected, params.σ_y), data.μ_y[i][2:end] ))
            
        end
        return LogDVal(ll)
    end
        
end

log_likelihood = log_lik()
```

```julia
posterior = PosteriorDensity(log_likelihood, prior_ang);
```

```julia
posterior_is = bat_transform(PriorToGaussian(), posterior, PriorSubstitution()).result

iters = 0000
iters_warmup = 600
chains = 5

metric = BAT.DiagEuclideanMetric()
integrator = BAT.LeapfrogIntegrator(0.0)

proposal = BAT.NUTS(:MultinomialTS, :ClassicNoUTurn)
# proposal = BAT.FixedTrajectoryLength()

adaptor = BAT.StanHMCAdaptor(0.8, iters_warmup)
# adaptor = BAT.StepSizeAdaptor()

 ahmc_sampler = HamiltonianMC(metric, ForwardDiff, integrator, proposal, adaptor);
```

```julia
function predict_position(x0, y0, θ, α; s = Float64[1.478, 15.026, 23.1150])
    x_expected = x0 .+ cos(α).*tan(θ*10^-5).*s.*10^3
    y_expected = y0 .+ sin(α).*tan(θ*10^-5).*s.*10^3
    return (x_expected, y_expected)
end
```
